```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv("pwt110.csv", delimiter=';')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
isnl = df.isnull().sum().sort_values(ascending=False)
print(isnl)
```

```python
df['country'].dtype
```

```python
sizes_cig = [['Extrapolated', 
             'Benchmark', 
             'Interpolated', 
             'ICP PPP timeseries: benchmark or interpolated', 
             'ICP PPP timeseries: extrapolated']]

df['i_cig'] = df['i_cig'].fillna('Extrapolated')  # or any valid category from sizes_cig   # same for i_xm

ordinal_encoder_cig = OrdinalEncoder(categories=sizes_cig)
df['i_cig_encoded'] = ordinal_encoder_cig.fit_transform(df[['i_cig']])

print(df[['i_cig', 'i_cig_encoded']].head())
```

```python
print(df['i_cig_encoded'].unique())
```

```python
sizes_xm = [['Extrapolated', 'Benchmark', 'Interpolated']]
df['i_xm'] = df['i_xm'].fillna('Extrapolated') 

ordinal_encoder_xm = OrdinalEncoder(categories=sizes_xm)
df['i_xm_encoded'] = ordinal_encoder_xm.fit_transform(df[['i_xm']])

print(df[['i_xm', 'i_xm_encoded']].head())
```

```python
print(df['i_xm_encoded'].unique())
```

```python
df.columns
```

```python
categorical_columns = ['i_xr', 'i_outlier', 'i_irr']
selected_columns = df[categorical_columns].select_dtypes(include=['object']).columns.tolist()
encoder = OneHotEncoder(sparse_output=False)

one_hot_encoded = encoder.fit_transform(df[selected_columns])

one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(selected_columns))

df_encoded = pd.concat([df, one_hot_df], axis=1)

df_encoded = df_encoded.drop(selected_columns, axis=1)
print(f"Encoded data : \n{df_encoded}")
```
